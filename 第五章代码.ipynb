```python
import requests
url = 'https://search.jd.com/search'
#以字典存储查询的关键词及属性
qrydata = {
    'keyword':'京东',
    'enc':'utf-8',
}
lt = []
for k,v in qrydata.items():
    lt.append(k+'='+str(v))
query_string = '&'.join(lt)
url = url + '?'+query_string
print(url)
r=requests.get(url)
print(r.status_code)
fb=open("C:\\Users\\Administrator\\Desktop\\爬虫\\京东.html","wb")
fb.write(r.content)
fb.close()
```
