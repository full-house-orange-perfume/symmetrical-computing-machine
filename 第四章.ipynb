```python
import requests
# 1. 核心请求头
headers = {
'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36',   
'Cookie': 'https://item.taobao.com/item.htm?spm=a219r.lm897.14.38.5d2346e28rO73l&id=522811099442&ns=1&abbucket=7', 
'Referer': 'https://www.taobao.com/',  
'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
'Accept-Language': 'zh-CN,zh;q=0.9' 
 }
# 2. 请求配置
url = 'https://www.taobao.com/'
timeout = 15  # 延长超时时间，应对网络波动
try: 
    # 3. 发送GET请求    
    response = requests.get(
        url,     
        headers=headers,
        timeout=timeout,
        verify=True  # 验证SSL证书，避免安全风险
  )   
# 4. 自动适配编码（解决中文乱码）   
    if response.encoding == 'ISO-8859-1':
       response.encoding = response.apparent_encoding  
       # 用chardet自动检测编码  
# 5. 验证请求与保存   
    if response.status_code == 200:   # 检查是否为正常首页（含"淘宝"关键词则说明未被拦截）       
       if '淘宝' in response.text:         
           with open('C:/Users/86151/Desktop/爬虫/taobao_homepage.html', 'w', encoding=response.encoding) as f:
               f.write(response.text)          
               print("爬取成功！已保存为 taobao_homepage.html")        
       else:          
           print("内容异常，可能仍被拦截")  
    else: 
        print(f"请求失败，状态码：{response.status_code}")
except requests.exceptions.ConnectTimeout:
    print("连接超时，请检查网络")
except requests.exceptions.SSLError:   
    print("SSL证书错误，可尝试将verify改为False")
except Exception as e:  
    print(f"其他错误：{str(e)}")
```
